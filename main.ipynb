```python
import numpy as np
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))

from forex_python.converter import CurrencyRates
c = CurrencyRates()
```

## Create vectors of time range

```python
month_range = [datetime.now() - timedelta(days=i) for i in range(30)]
year_range = [datetime.now() - timedelta(days=i) for i in range(365)]
years5_range = [datetime.now() - timedelta(days=i) for i in range(365*5)]
```

## Get exchange rates

```python
month_mean_rates = {'USD': [], 'EUR': [], 'GBP': [], 'CHF': []}
month_pln_rates = {'USD': [], 'EUR': [], 'GBP': [], 'CHF': []}

year_mean_rates = {'USD': [], 'EUR': [], 'GBP': [], 'CHF': []}
year_pln_rates = {'USD': [], 'EUR': [], 'GBP': [], 'CHF': []}

years5_mean_rates = {'USD': [], 'EUR': [], 'GBP': [], 'CHF': []}
years5_pln_rates = {'USD': [], 'EUR': [], 'GBP': [], 'CHF': []}
```

#### Month

```python
for day in month_range:
    # USD
    rates_dict = c.get_rates('USD', day)
    month_mean_rates['USD'].append(np.average([rates_dict['EUR'], rates_dict['GBP'], rates_dict['CHF']]))
    month_pln_rates['USD'].append(rates_dict['PLN'])

    # EUR
    rates_dict = c.get_rates('EUR', day)
    month_mean_rates['EUR'].append(np.average([rates_dict['USD'], rates_dict['GBP'], rates_dict['CHF']]))
    month_pln_rates['EUR'].append(c.get_rates('EUR', day)['PLN'])
    
    # GBP
    rates_dict = c.get_rates('GBP', day)
    month_mean_rates['GBP'].append(np.average([rates_dict['USD'], rates_dict['EUR'], rates_dict['CHF']]))
    month_pln_rates['GBP'].append(c.get_rates('GBP', day)['PLN'])
    
    # CHF
    rates_dict = c.get_rates('CHF', day)
    month_mean_rates['CHF'].append(np.average([rates_dict['USD'], rates_dict['EUR'], rates_dict['GBP']]))
    month_pln_rates['CHF'].append(c.get_rates('CHF', day)['PLN'])
```

#### Year

```python
for day in year_range:
    # USD
    rates_dict = c.get_rates('USD', day)
    year_mean_rates['USD'].append(np.average([rates_dict['EUR'], rates_dict['GBP'], rates_dict['CHF']]))
    year_pln_rates['USD'].append(rates_dict['PLN'])

    # EUR
    rates_dict = c.get_rates('EUR', day)
    year_mean_rates['EUR'].append(np.average([rates_dict['USD'], rates_dict['GBP'], rates_dict['CHF']]))
    year_pln_rates['EUR'].append(c.get_rates('EUR', day)['PLN'])
    
    # GBP
    rates_dict = c.get_rates('GBP', day)
    year_mean_rates['GBP'].append(np.average([rates_dict['USD'], rates_dict['EUR'], rates_dict['CHF']]))
    year_pln_rates['GBP'].append(c.get_rates('GBP', day)['PLN'])
    
    # CHF
    rates_dict = c.get_rates('CHF', day)
    year_mean_rates['CHF'].append(np.average([rates_dict['USD'], rates_dict['EUR'], rates_dict['GBP']]))
    year_pln_rates['CHF'].append(c.get_rates('CHF', day)['PLN'])
```

#### 5 years

```python
for day in years5_range:
    # USD
    rates_dict = c.get_rates('USD', day)
    years5_mean_rates['USD'].append(np.average([rates_dict['EUR'], rates_dict['GBP'], rates_dict['CHF']]))
    years5_pln_rates['USD'].append(rates_dict['PLN'])

    # EUR
    rates_dict = c.get_rates('EUR', day)
    years5_mean_rates['EUR'].append(np.average([rates_dict['USD'], rates_dict['GBP'], rates_dict['CHF']]))
    years5_pln_rates['EUR'].append(c.get_rates('EUR', day)['PLN'])
    
    # GBP
    rates_dict = c.get_rates('GBP', day)
    years5_mean_rates['GBP'].append(np.average([rates_dict['USD'], rates_dict['EUR'], rates_dict['CHF']]))
    years5_pln_rates['GBP'].append(c.get_rates('GBP', day)['PLN'])
    
    # CHF
    rates_dict = c.get_rates('CHF', day)
    years5_mean_rates['CHF'].append(np.average([rates_dict['USD'], rates_dict['EUR'], rates_dict['GBP']]))
    years5_pln_rates['CHF'].append(c.get_rates('CHF', day)['PLN'])
```

## Visualize

### Month

#### USD

```python
fig, ax1 = plt.subplots(figsize=(16, 6))
ax2 = ax1.twinx()
ax1.plot(month_range, month_mean_rates['USD'], label='mean', color='blue')
ax2.plot(month_range, month_pln_rates['USD'], label='PLN', color='green')

ax1.set_xlabel('Days', fontsize=16)
ax1.set_ylabel('Średnia cena (EUR, GBP, CHF)', color='blue', fontsize=18)
ax2.set_ylabel('Cena w PLN', color='green', fontsize=18)

ax1.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='major', labelsize=12)

plt.title("Średnia cena dolara (EUR, GBP, CHF) oraz cena w PLN", fontsize=20)
plt.grid()

plt.show()
```
