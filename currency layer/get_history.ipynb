```python
import random
all_results = []

for date_ in dates_range:
    api_key = api_keys[random.randint(0, 1)]
    url = "http://api.currencylayer.com/historical?access_key=" + api_key + "&date=" + date_ + "&currencies=" + cur_list + "&format=1"
    resp = requests.get(url).json()
    
    res_dict = {resp["date"]: resp["quotes"]}
    all_results.append(res_dict)
```

```python
with open("caly_2016.json", "w") as f:
    json.dump(all_results, f)
```
