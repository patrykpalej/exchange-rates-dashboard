```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))

import json
import requests
from datetime import date, timedelta
```

```python
api_key = "c6306de007a4cf317334f9cd73156c53"  # palej@agh.edu.pl
cur_list = "EUR,GBP,CHF,PLN"
dates_range = [(date(2020, 4, 1) + timedelta(days=i)).strftime("%Y-%m-%d") for i in range(180)]
```

```python
dates_range
```

```python
all_results = []

for date_ in dates_range:
    url = "http://api.currencylayer.com/historical?access_key=" + api_key + "&date=" + date_ + "&currencies=" + cur_list + "&format=1"
    resp = requests.get(url).json()
    
    res_dict = {resp["date"]: resp["quotes"]}
    all_results.append(res_dict)
```

```python
with open("results.json", "w") as f:
    json.dump(all_results, f)
```
