```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))

import json
import pandas as pd
from datetime import date, timedelta, datetime
```

```python
with open("results_19.json", "r") as f:
    res = json.load(f)
```

```python
cols = pd.read_csv('../data/exchange_rates.csv', sep=';').columns
```

```python
df = pd.DataFrame(columns=cols)

for day in res:
    timestamp = datetime.strptime(list(day.keys())[0], "%Y-%m-%d").strftime("%d.%m.%Y")
    val = day[list(day.keys())[0]]
    
    usdeur = val['USDEUR']
    usdgbp = val['USDGBP']
    usdchf = val['USDCHF']
    
    eurusd = 1 / val['USDEUR']
    eurgbp = val['USDGBP'] / val['USDEUR']
    eurchf = val['USDCHF'] / val['USDEUR']
    
    gbpusd = 1 / val['USDGBP']
    gbpeur = val['USDEUR'] / val['USDGBP']
    gbpchf = val['USDCHF'] / val['USDGBP']
    
    chfusd = 1 / val['USDCHF']
    chfeur = val['USDEUR'] / val['USDCHF']
    chfgbp = val['USDGBP'] / val['USDCHF']
    
    usdpln = val['USDPLN']
    eurpln = val['USDPLN'] / val['USDEUR']
    gbppln = val['USDPLN'] / val['USDGBP']
    chfpln = val['USDPLN'] / val['USDCHF']
    
    df.loc[len(df)] = [timestamp, usdeur, usdgbp, usdchf, eurusd, eurgbp, eurchf, gbpusd, gbpeur, gbpchf, chfusd, chfeur, chfgbp, usdpln, eurpln, gbppln, chfpln]
```
